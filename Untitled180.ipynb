## Load or simulate dataset

### Subtask:
Load a real dataset (e.g., MovieLens) or simulate a dataset with 'runtime', 'rating', and binary genre indicators.


**Reasoning**:
Simulate the movie dataset with runtime, rating, and genre indicators, create a DataFrame, and display its head and info.



```python
import numpy as np
import pandas as pd

# Simulate data
np.random.seed(42) # for reproducibility
n_samples = 200
runtime = np.random.randint(60, 181, size=n_samples)
rating = np.random.uniform(1.0, 10.0, size=n_samples)

genres = ['action', 'comedy', 'drama', 'sci-fi', 'thriller']
genre_data = {}
for genre in genres:
    genre_data[f'genre_{genre}'] = np.random.randint(0, 2, size=n_samples)

data = {'runtime': runtime, 'rating': rating, **genre_data}
df = pd.DataFrame(data)

# Display first few rows and info
display(df.head())
df.info()
```

## Exploratory data analysis (eda)

### Subtask:
Perform EDA, including plotting a correlation heatmap of features using seaborn.


**Reasoning**:
Calculate the correlation matrix and create a heatmap to visualize the correlations between features in the DataFrame.



```python
correlation_matrix = df.corr()

import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Feature Correlation Heatmap')
plt.show()
```

## Data preprocessing

### Subtask:
Standardize all columns using StandardScaler.


**Reasoning**:
Standardize the features in the DataFrame using StandardScaler.



```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
df_scaled = scaler.fit_transform(df)
df_scaled = pd.DataFrame(df_scaled, columns=df.columns)

display(df_scaled.head())
```

## K-means clustering

### Subtask:
Apply KMeans clustering with 4 clusters and add cluster labels to the DataFrame.


**Reasoning**:
Apply KMeans clustering to the scaled data and add the cluster labels to the original dataframe.



```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=4, random_state=42, n_init=10)
kmeans.fit(df_scaled)
cluster_labels = kmeans.labels_
df['cluster_label'] = cluster_labels

display(df.head())
```

## Cluster profiling

### Subtask:
Group data by cluster and calculate mean values for all features, and display results as a heatmap.


**Reasoning**:
Group the DataFrame by cluster labels, calculate the mean of features for each cluster, and then visualize these mean values as a heatmap to profile the clusters.



```python
cluster_profiles = df.groupby('cluster_label').mean()

import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
sns.heatmap(cluster_profiles, annot=True, cmap='YlGnBu', fmt=".2f")
plt.title('Cluster Profile Heatmap (Mean Feature Values)')
plt.show()
```

## Dimensionality reduction (optional)

### Subtask:
Use PCA to project data into 2D space.


**Reasoning**:
The next step is to perform PCA on the scaled data to reduce its dimensionality for visualization.



```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2, random_state=42)
df_pca = pca.fit_transform(df_scaled)
df_pca = pd.DataFrame(df_pca, columns=['PC1', 'PC2'])

display(df_pca.head())
```

## Visualize clusters

### Subtask:
Plot clusters in a 2D scatter plot with distinct colors.


**Reasoning**:
Plot the 2D PCA reduced data points, colored by their cluster labels, to visualize the clusters in a 2D space with appropriate labels, title, and a legend.



```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 8))
scatter = plt.scatter(df_pca['PC1'], df_pca['PC2'], c=df['cluster_label'], cmap='viridis', s=50, alpha=0.7)
plt.title('K-Means Clusters in 2D PCA Space')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.colorbar(scatter, label='Cluster Label')
plt.grid(True)
plt.show()
```

```python
# Display the feature correlation heatmap (previously generated)
print("Feature Correlation Heatmap:")
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Feature Correlation Heatmap')
plt.show()

# Display the cluster profile heatmap (previously generated)
print("\nCluster Profile Heatmap (Mean Feature Values):")
plt.figure(figsize=(12, 6))
sns.heatmap(cluster_profiles, annot=True, cmap='YlGnBu', fmt=".2f")
plt.title('Cluster Profile Heatmap (Mean Feature Values)')
plt.show()

# Display the 2D cluster scatter plot (previously generated)
print("\nK-Means Clusters in 2D PCA Space:")
plt.figure(figsize=(10, 8))
scatter = plt.scatter(df_pca['PC1'], df_pca['PC2'], c=df['cluster_label'], cmap='viridis', s=50, alpha=0.7)
plt.title('K-Means Clusters in 2D PCA Space')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.colorbar(scatter, label='Cluster Label')
plt.grid(True)
plt.show()

# Provide a summary interpreting the clusters
print("\nCluster Interpretation Summary:")
print("Based on the Cluster Profile Heatmap:")
print("Cluster 0: Characterized by high action and sci-fi genre prevalence, moderate runtime, and average rating. These could be action-packed sci-fi movies.")
print("Cluster 1: Characterized by high thriller, sci-fi, and comedy genre prevalence, slightly above-average runtime, but below-average rating. These might be longer, less well-received thriller/sci-fi comedies.")
print("Cluster 2: Characterized by moderate comedy, drama, and sci-fi genre prevalence, and slightly above-average runtime and rating. These could be well-regarded comedy-drama/sci-fi films.")
print("Cluster 3: Characterized by high drama and thriller genre prevalence, slightly above-average runtime, but very low rating. These might be poorly-rated dramatic thrillers.")
```
